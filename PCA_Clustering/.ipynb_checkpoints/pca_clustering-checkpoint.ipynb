# <a id='3.0'>3. Dimensionality Reduction (Principal Components Analysis)</a>

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
RANDOM_STATE = 42

train_set, test_set = train_test_split(data_df, test_size=0.25, random_state=RANDOM_STATE)
```

```python
%%time
n_components_range = [int(x) for x in range(1, len(data_df.columns)+1)]
for n_components in n_components_range:
    scaler = StandardScaler()
    pca = PCA(n_components=n_components)
    pipe = Pipeline([('scaler', scaler), ('pca', pca)])
    pipe.fit(train_set)
    print(f'{n_components} components explains',
          list(map('{:.1f}%'.format,pipe.steps[1][1].explained_variance_ratio_.cumsum()*100)).pop(),
          'of total variance')
```

```python
plt.figure(figsize=(16,9))
plt.title('Total Variance Explained Method')
plt.ylabel('Cumulative Variance Explained')
plt.xlabel('Principal Component')
plt.axvline(6)
plt.axhline(0.761)
plt.plot(np.arange(1, pca.n_components_ + 1),
         pipe.steps[1][1].explained_variance_ratio_.cumsum(),
         'ro-')
```

```python
plt.figure(figsize=(16,9))
plt.title('Scree Plot (Elbow & Eigenvalue Method)')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.plot(np.arange(1, pca.n_components_ + 1),
         pca.explained_variance_,
         'ro-',
         linewidth=2)
plt.axhline(1)
```

Cumulative explained variance, Eigenvalue, and Elbow methods all suggest **6 components** to be optimal (and especially for parsimonious reasons).

```python
chosen_number = 6
pca = PCA(n_components=chosen_number)
pipe = Pipeline([('scaler', scaler), ('pca', pca)])
fit = pipe.fit(train_set)
pc_df = pd.DataFrame(pipe.steps[1][1].components_, 
                     columns=[col + '_Z' for col in data_df.columns],
                     index=['PC_' + str(x) for x in range(1,chosen_number+1)]).T
pc_df_sig = pc_df
for column in pc_df_sig.columns:
    pc_df_sig.loc[(abs(pc_df_sig[column]) < 0.3), str(column)] = np.NaN
pc_df_sig
```

```python
# Verify Principal Components are not correlated
plt.figure(figsize =(10,10))
sns.heatmap(pd.DataFrame(pipe.fit_transform(train_set), columns=['PC_' + str(x) for x in range(1,7)]).corr(),
           annot=True,
           cmap='vlag')
```

```python
# Validate against Test Set
chosen_number = 6
pca_test = PCA(n_components=chosen_number)
pipe_test = Pipeline([('scaler', scaler), ('pca', pca)])
fit_test = pipe_test.fit(test_set)
pc_df_test = pd.DataFrame(pipe_test.steps[1][1].components_, 
                     columns=[col + '_Z' for col in data_df.columns],
                     index=['PC_' + str(x) for x in range(1,chosen_number+1)]).T
pc_df_test_sig = pc_df_test
for column in pc_df_test_sig.columns:
    pc_df_test_sig.loc[(abs(pc_df_test_sig[column]) < 0.3), str(column)] = np.NaN
pc_df_test_sig
```

The weights, and correspondance, on the test set are close, although not exact, to the weights on the training set. Validation successful.

## <a id='3.1'>3.1 Interpretation of Principal Components</a>

The trade-off with reducing the dimensionality of the dataset so that "redundant" variables are excluded from a supervised learning algorithm (or an unsupervised algorithm) is that the resulting dimensions, after a dimensionality reduction technique (such as principal components), of the dataset can be difficult to interpret. A trade-off between interpretability and algorithm efficiency.

I shall attempt to interpret the principal components on the training set above (although not all maybe interpretable).

* Component 1: Weights biggest along variables related to purchases (amount, count, frequency). **Purchases component**.
* Component 2: Strongly associated with Cash Advance variables and Balance. **Cash Advance and Balance component**.
* Component 3: Related to frequency variables. **Activity Component**.
* Component 5: Weights on one-off purchases and installments. **One-off & Installments component**.

Component 4 and 6 are a little more difficult to interpret as there is much overlap (on size and direction of weights). That said, there are some unique weights to each component.

* Component 4: PRC_FULL_PAYMENT & MINIMUM_PAYMENTS are largest absolute weights (each inverse to each other). **Payment component**.
* Component 6: Credit limit is the unique weight in this component (strongly associated with Tenured variable, as expected). **Credit Limit component**.

**Dataset has been reduced from 17 variables (dimensions) to 6 variables (dimensions)**

# <a id='4.0'>4. Clustering</a>

Clustering, like dimensionality reduction, is an example of *unsupervised learning* (where data is unlabelled, there is no "target"). 

While dimensionality reduction seeks to group similar dimensiosn so as to reduce the features in a dataset, clustering seeks to group similar *observations* (instances) by creating one or more dimensions to group similar records together.

## <a id='4.1'>4.1 Choice of Clustering Algorithm</a>

Many clustering algorithms exist and each is suited to a different purpose. For example, ***hierarchical clustering*** is useful to describe segments which have structure (one segment is the child/parent of another segment). An example of hierarchical clustering would be biological taxonomy where Mammals is a parent cluster with child clusters of Primates, Rodents, etc. This type of clustering is suited for when hierarchy is useful.

Another clustering algorithm is ***k-means (or k-medians) clustering*** which seeks to find k different clusters within the data and assign each observation to its nearest cluster (via a distance measure) so that each observation belongs to one (and only one) cluster (no hierachy is formed). This algorithm is suitable for the problem at hand since each observation represents a customer, I would like to understand if there are data-driven clusters (segments) of customers with similar spending behaviours.

## <a id='4.2'>4.2 k-means Clustering</a>

K-means clustering has some important features that need considering:

1. As k-means clustering minimizes within-cluster variances (squared Euclidean distances), it helps if the variables used have similar variances. This means standardizing the variable set yields "better" clusters.
2. k can range from 1 (all observations belong to one cluster) to the number of observations in the dataset (each observation is its own cluster) - the optimal k is lies within this range. To find the optimal number of clusters, [silhouette scores](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html#sklearn.metrics.silhouette_score) can be calculated for a range of k. The k with a silhouette score closest to 1 is considered "optimal".

For the first consideration, this has already been achieved through the principal components exercise above, so I only need to consider the appropriate value of k.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
%%time
range_n_clusters = [int(x) for x in range(2,11)]

inertia_score_list = []
silhouette_score_list = []
for n_clusters in range_n_clusters:    
    scaler = StandardScaler()
    pca = PCA(n_components=6)
    clusterer = KMeans(n_clusters=n_clusters, random_state=RANDOM_STATE)
    pipe = Pipeline([('scaler', scaler), ('pca', pca), ('cluster',clusterer)])
    cluster_labels = pipe.fit_predict(data_df)
    
    inertia_score_list.append(pipe['cluster'].inertia_)
    silhouette_score_list.append(silhouette_score(data_df, cluster_labels))
    
    print(f"For n_clusters = {n_clusters}, the average silhouette score is :{silhouette_score_list[n_clusters-2]:.2f} and the inertia is {inertia_score_list[n_clusters-2]:,.1f}")
```

```python
# Elbow Method
plt.figure(figsize=(16,9))
plt.title('Inertia Score Method (Elbow Method)')
plt.xlabel('k Clusters')
plt.ylabel('Inertia')
plt.plot(np.arange(2, len(inertia_score_list)+2),
         inertia_score_list,
         'ro-',
         linewidth=2)
```

```python
# Silhouette Score Method
plt.figure(figsize=(16,9))
plt.title('Silhouette Score Method')
plt.xlabel('k Clusters')
plt.ylabel('Silhouette Score')
plt.plot(np.arange(2, len(inertia_score_list)+2),
         silhouette_score_list,
         'ro-',
         linewidth=2)
```

The Elbow method suggests 4 clusters and the Silhouette method suggests 3 clusters. For parsimonious reasons , 3 clusters is more preferable to 4 clusters. I will visualise a sample of the data for these three clusters along the two first principal components.

```python
scaler = StandardScaler()
pca = PCA(n_components=6)
clusterer = KMeans(n_clusters=3, random_state=RANDOM_STATE)
pipe1 = Pipeline([('scaler', scaler), ('pca', pca)])
pipe2 = Pipeline([('pipe1',pipe1),('cluster',clusterer)])
cluster_labels = pipe2.fit_predict(data_df)
```

```python
pc_data_df = pd.DataFrame(pipe1.fit_transform(data_df),columns=['PC_' + str(x) for x in range(1,7)])
cluster_df = pd.DataFrame(cluster_labels, columns=['cluster'])
final_df = pd.concat([pc_data_df,cluster_df],axis=1)
final_df.head()
```

## <a id='4.3'>4.3 Describing the Clusters</a>

The clusters are described by their centroids across the reduced (via principal components) dimensions.

```python
pd.options.display.float_format = '{:,.2f}'.format
cluster_centers = pd.DataFrame(pipe2['cluster'].cluster_centers_, 
                               index=['Cluster_' + str(x) for x in range(1,4)],
                               columns=['PC_' + str(x) for x in range(1,7)])
print(cluster_centers.T)
```

* Cluster 1: Large positive mean on PC 2 (corresponding to Cash Advance and Balance) -> **Customers with large cash advances and balances**.
* Cluster 2: Negative means on PC 1 and PC 2 -> **Low value customers**.
* Cluster 3: Large positive mean on PC 1 (corresponding to Purcahses) -> **Customers with highest purchasing activity**.

## <a id='4.4'>4.4 Visualising the Clusters</a>

```python
colors = ['#4EACC5', '#FF9C34', '#4E9A06']
PC_Means = sample_final.groupby('cluster')[['PC_1','PC_2']].mean().reset_index()

plt.figure(figsize=(16,9))
plt.title('Cluster Visualisation')
plt.xlabel('Princiapl Component 1')
plt.ylabel('Princiapl Component 2')
for k, col in zip(range(3), colors):
    plt.plot(sample_final.loc[sample_final['cluster']==k]['PC_1'],
            sample_final.loc[sample_final['cluster']==k]['PC_2'],
            'w',
            markerfacecolor=col,
            marker='.',
            markersize=20)
    plt.plot(PC_Means.loc[PC_Means['cluster']==k]['PC_1'],
             PC_Means.loc[PC_Means['cluster']==k]['PC_2'],
             'o',
             markerfacecolor=col,
             markeredgecolor='k',
             markersize=15)
```
